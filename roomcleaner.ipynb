```python
import random

class RoomCleanerAgent:
    def __init__(self, room_size=(10, 10)):
        self.room_size = room_size
        # Create a room where 1 represents a dirty cell and 0 represents a clean cell
        self.grid = [[random.choice([0, 1]) for _ in range(room_size[1])] for _ in range(room_size[0])]
        # Starting position is chosen randomly within the room bounds
        self.current_position = (random.randint(0, room_size[0] - 1), random.randint(0, room_size[1] - 1))

    def display_room(self):
        # Display the room grid, with 0 representing clean and 1 representing dirty
        print("\nCurrent Room Status:")
        for row in self.grid:
            print(" ".join(map(str, row)))
        print()

    def clean(self):
        # Clean the current position (set it to 0, which represents clean)
        x, y = self.current_position
        if self.grid[x][y] == 1:
            print(f"Cleaning position {self.current_position} (dirty).")
            self.grid[x][y] = 0
        else:
            print(f"Position {self.current_position} is already clean.")

    def move(self):
        # Move randomly in one of the four possible directions (up, down, left, right)
        x, y = self.current_position
        direction = random.choice(['up', 'down', 'left', 'right'])

        if direction == 'up' and x > 0:
            x -= 1
        elif direction == 'down' and x < self.room_size[0] - 1:
            x += 1
        elif direction == 'left' and y > 0:
            y -= 1
        elif direction == 'right' and y < self.room_size[1] - 1:
            y += 1

        self.current_position = (x, y)
        print(f"Moving to position {self.current_position}.")

    def is_room_clean(self):
        # Check if the entire room is clean (all cells should be 0)
        return all(cell == 0 for row in self.grid for cell in row)

    def run(self, steps=10):
        print("Initial Room Status:")
        self.display_room()

        # Run the agent for a number of steps or until the room is fully clean
        step_count = 0
        while step_count < steps and not self.is_room_clean():
            print(f"Step {step_count + 1}:")
            print(f"Agent is at position {self.current_position}")
            self.clean()  # Clean the current position
            self.move()   # Move to a new position
            self.display_room()  # Display the updated room
            step_count += 1

        if self.is_room_clean():
            print("Room is fully cleaned!")
        else:
            print(f"Completed {step_count} steps, but the room is not fully cleaned.")

# Example usage
agent = RoomCleanerAgent(room_size=(3, 3))  # Initialize a smaller room for demonstration
agent.run(steps=10)  # Run the agent for 10 steps
```
